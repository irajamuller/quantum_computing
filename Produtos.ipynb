```python
import numpy as np
from qiskit.visualization import plot_bloch_multivector
import qiskit.quantum_info as qi
# plot_bloch_multivector(qi.Statevector(psi))
```

```python
# Produto interno
ket = np.array([[1], [0]])
bra = ket.T
bra
```

```python
print(bra @ ket) # <0|0>
```

```python
ket1 = np.array([[0], [1]])
bra1 = ket1.T
print(bra1 @ ket1) # <1|1>
```

```python
print(bra @ ket1) # <0|1>
print(bra1 @ ket) # <1|0>
```

```python
# Produto externo (projetores)
print(ket @ bra) # |0><0|
print(ket @ bra1) # |0><1|
print(ket1 @ bra) # |1><0|
print(ket1 @ bra1) # |1><1|
```

```python
# Produto Tensorial
print(np.kron(ket, ket)) # |00>
print(np.kron(ket, ket1)) # |01>
print(np.kron(ket1, ket)) # |10>
print(np.kron(ket1, ket1)) # |11>
```

```python
alpha = 1/np.sqrt(2)
beta = 1/np.sqrt(2)
psi = alpha * ket + beta * ket1
plot_bloch_multivector(qi.Statevector(psi))
```

```python
alpha = 1
beta = 0
psi = alpha * ket + beta * ket1
plot_bloch_multivector(qi.Statevector(psi))
```

```python
X = np.array([ [0, 1], [1, 0]])
```

```python
psi2 = X @ psi
print(psi2)
plot_bloch_multivector(qi.Statevector(psi2))
```

```python
H = 1/np.sqrt(2) * np.array([ [1, 1], [1, -1]])
```

```python
psi3 = H @ psi
print(psi3)
plot_bloch_multivector(qi.Statevector(psi3))
```

```python
dois = np.kron(H @ ket1, np.kron(ket1, ket1))
print(dois)
plot_bloch_multivector(qi.Statevector(dois))
```
